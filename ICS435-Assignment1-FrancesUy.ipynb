## Imports

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

## Dataset
_We will be using the **Breast Cancer dataset** provided by sklearn. It includes 30 features and a binary classification task (malignant vs. benign)_

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

print("Available attributes:", data.keys())
print("\nFirst two rows of measurements:")
print(data.data[:2])

# from available attributes, the features is accessed with 'data'
# and output is acesssed with 'target'
X = data.data
y = data.target
```

_Verifying our data loaded correctly. This means there are 569 samples (patients), 30 features for each sample (patient), and there are 569 possible binary outputs for each one (malignant vs. benign)_

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Data Preprocessing

_Partition the data into an **80% training set** and a **20% test set**_

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

_Verifying our split sizes_

```python
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

_Scale the features using **StandardScaler** for KNN_

```python
scaler = preprocessing.StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

***

## K-Nearest Neighbors (KNN)

```python
# train KNN model with n_neighbors=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# make predictions
y_pred_knn = knn.predict(X_test_scaled)

# calculate metrics
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)

# print model results
print("KNN Model Results:")
print("======================")
print(f"Accuracy: {accuracy_knn:.4f}")
print(f"Precision: {precision_knn:.4f}")
print(f"Recall: {recall_knn:.4f}")
print(f"F1-score: {f1_knn:.4f}")



# confusion matrix plot with numbers using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_knn, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - KNN')
plt.ylabel('True label')
plt.xlabel('Predicted label')

# add labels for clarity
labels = ['Benign (0)', 'Malignant (1)']
plt.xticks([0.5, 1.5], labels)
plt.yticks([0.5, 1.5], labels, rotation=0)

plt.show()

# ablation study for KNN
n_neighbors_range = [3, 5, 7, 9, 11, 13, 15]
accuracy_scores = []

for n in n_neighbors_range:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    
# plot ablation study results
plt.figure(figsize=(10, 6))
plt.plot(n_neighbors_range, accuracy_scores, marker='o')
plt.title('KNN Performance vs Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

***

## Decision Tree

```python
# train Decision Tree model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# make predictions
y_pred_dt = dt.predict(X_test)

# calculate metrics
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)

# print model results
print("Decision Tree Model Results:")
print("======================")
print(f"Accuracy: {accuracy_dt:.4f}")
print(f"Precision: {precision_dt:.4f}")
print(f"Recall: {recall_dt:.4f}")
print(f"F1-score: {f1_dt:.4f}")

# confusion matrix plot with numbers using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_dt, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Decision Tree')
plt.ylabel('True label')
plt.xlabel('Predicted label')


# ablation study for Decision Tree
max_depth_range = [3, 5, 7, 10, 15, 20, None]
accuracy_scores_dt = []

for depth in max_depth_range:
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    accuracy_scores_dt.append(accuracy_score(y_test, y_pred))

# plot ablation study results
plt.figure(figsize=(10, 6))
plt.plot([str(d) for d in max_depth_range], accuracy_scores_dt, marker='o')
plt.title('Decision Tree Performance vs Max Depth')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

***

## Random Forest

```python
# train Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# make predictions
y_pred_rf = rf.predict(X_test)

# calculate metrics
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

# print model results
print("Random Forest Model Results:")
print("======================")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-score: {f1_rf:.4f}")

# confusion matrix plot with numbers using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest')
plt.ylabel('True label')
plt.xlabel('Predicted label')


# ablation studies for Random Forest
max_depth_range = [3, 5, 7, 10, 15, 20, None]
min_samples_split_range = [2, 5, 10, 20, 50]

# test max_depth
accuracy_scores_rf_depth = []
for depth in max_depth_range:
    rf = RandomForestClassifier(n_estimators=100, max_depth=depth, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    accuracy_scores_rf_depth.append(accuracy_score(y_test, y_pred))

plt.figure(figsize=(10, 6))
plt.plot([str(d) for d in max_depth_range], accuracy_scores_rf_depth, marker='o')
plt.title('Random Forest Performance vs Max Depth')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# test min_samples_split
accuracy_scores_rf_split = []
for split in min_samples_split_range:
    rf = RandomForestClassifier(n_estimators=100, min_samples_split=split, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    accuracy_scores_rf_split.append(accuracy_score(y_test, y_pred))

plt.figure(figsize=(10, 6))
plt.plot(min_samples_split_range, accuracy_scores_rf_split, marker='o')
plt.title('Random Forest Performance vs Min Samples Split')
plt.xlabel('Min Samples Split')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

***

## Model Comparison

```python
# create comparison table
results_df = pd.DataFrame({
    'Model': ['KNN', 'Decision Tree', 'Random Forest'],
    'Accuracy': [accuracy_knn, accuracy_dt, accuracy_rf],
    'Precision': [precision_knn, precision_dt, precision_rf],
    'Recall': [recall_knn, recall_dt, recall_rf],
    'F1-score': [f1_knn, f1_dt, f1_rf]
})

print("\nModel Comparison:")
print(results_df.to_string(index=False))

# plot comparison bar chart
plt.figure(figsize=(12, 6))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-score']
x = np.arange(len(metrics))
width = 0.25

plt.bar(x - width, results_df.iloc[0, 1:], width, label='KNN')
plt.bar(x, results_df.iloc[1, 1:], width, label='Decision Tree')
plt.bar(x + width, results_df.iloc[2, 1:], width, label='Random Forest')

plt.xlabel('Metrics')
plt.ylabel('Score')
plt.title('Model Comparison Across Metrics')
plt.xticks(x, metrics)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
